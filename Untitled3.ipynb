```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC  # Импорт SVM
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.datasets import mnist

# Загрузка данных
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Предобработка данных
X_train = X_train / 255.0
X_test = X_test / 255.0

# Преобразование данных
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

# Разделение данных на обучающий и тестовый наборы
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Визуализация данных
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i, ax in enumerate(axes.ravel()):
    ax.imshow(X_train[i].reshape(28, 28), cmap='gray')
    ax.axis('off')
    ax.set_title(str(y_train[i]))
plt.show()

# Модель 1: RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)

# Модель 2: Support Vector Machine (SVM)
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

# Модель 3: LogisticRegression
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train, y_train)

# Оценка точности на валидационном наборе
rf_val_predictions = rf_model.predict(X_val)
svm_val_predictions = svm_model.predict(X_val)
lr_val_predictions = lr_model.predict(X_val)

rf_accuracy = accuracy_score(y_val, rf_val_predictions)
svm_accuracy = accuracy_score(y_val, svm_val_predictions)
lr_accuracy = accuracy_score(y_val, lr_val_predictions)

print(f'RandomForestClassifier Accuracy on validation set: {rf_accuracy:.4f}')
print(f'Support Vector Machine Accuracy on validation set: {svm_accuracy:.4f}')
print(f'LogisticRegression Accuracy on validation set: {lr_accuracy:.4f}')

# Оценка точности на тестовом наборе
rf_test_predictions = rf_model.predict(X_test)
svm_test_predictions = svm_model.predict(X_test)
lr_test_predictions = lr_model.predict(X_test)

rf_test_accuracy = accuracy_score(y_test, rf_test_predictions)
svm_test_accuracy = accuracy_score(y_test, svm_test_predictions)
lr_test_accuracy = accuracy_score(y_test, lr_test_predictions)

print(f'RandomForestClassifier Accuracy on test set: {rf_test_accuracy:.4f}')
print(f'Support Vector Machine Accuracy on test set: {svm_test_accuracy:.4f}')
print(f'LogisticRegression Accuracy on test set: {lr_test_accuracy:.4f}')

# Визуализация работы моделей
def plot_confusion_matrix(y_true, y_pred, classes, title='Confusion matrix', cmap=plt.cm.Blues):
    cm = confusion_matrix(y_true, y_pred)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Визуализация матрицы ошибок для RandomForestClassifier
plt.figure(figsize=(8, 6))
plot_confusion_matrix(y_test, rf_test_predictions, classes=np.arange(10), title='RandomForestClassifier Confusion Matrix')
plt.show()

# Визуализация матрицы ошибок для Support Vector Machine (SVM)
plt.figure(figsize=(8, 6))
plot_confusion_matrix(y_test, svm_test_predictions, classes=np.arange(10), title='SVM Confusion Matrix')
plt.show()

# Визуализация матрицы ошибок для LogisticRegression
plt.figure(figsize=(8, 6))
plot_confusion_matrix(y_test, lr_test_predictions, classes=np.arange(10), title='LogisticRegression Confusion Matrix')
plt.show()
```
